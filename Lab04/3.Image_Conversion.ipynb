```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread("images/abd.jpg")

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display grayscale image
plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Save grayscale image
cv2.imwrite("images/grayscale_image.jpg", gray)

print("Grayscale image saved successfully")
```

```python
# Save image in BMP format
cv2.imwrite("images/grayscale_image.bmp", gray)

print("JPG image successfully converted to BMP format")
```

```python
# Binary thresholding
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Display binary image
plt.imshow(binary, cmap="gray")
plt.title("Binary Image")
plt.axis("off")
plt.show()
```

```python
# Save binary image
cv2.imwrite("images/binary_image.jpg", binary)

print("Binary image saved successfully")
```

```python
# Save image in BMP format
cv2.imwrite("images/binary_image.bmp", binary)

print("JPG image successfully converted to BMP format")
```
